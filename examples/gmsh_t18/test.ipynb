```python
# elm = sg.mesh.cells[0][0]
# print(f'{elm=}')

# --- Node coordinates (replace these with your own) ---
# Example: four nodes in 3D space
# nodes = np.array([
#     [0.0, 0.0, 0.0],   # Node 0
#     [1.0, 0.0, 0.0],   # Node 1
#     [0.5, 0.8, 0.0],   # Node 2
#     [0.3, 0.4, 1.0]    # Node 3
# ])
nodes = []
nid_to_print = [297, 288, 320, 275]
for nid in nid_to_print:
    nodes.append(sg.mesh.points[nid])
    # print(f'{sg.mesh.points[nid-1]=}')
nodes = np.array(nodes)
print(f'{nodes=}')

# --- Connectivity: one tetrahedron (4 nodes) ---
# PyVista expects: [n_pts, p0, p1, p2, p3]
# where n_pts = number of points in the cell (here, 4)
tetra = np.hstack([[4], np.arange(4)])

# --- Create the unstructured grid ---
cells = np.array([tetra], dtype=np.int64)
cell_types = np.array([pv.CellType.TETRA], dtype=np.uint8)

grid = pv.UnstructuredGrid(cells, cell_types, nodes)

# --- Plotting ---
plotter = pv.Plotter()

# Add the tetrahedral mesh
plotter.add_mesh(grid, show_edges=True, color="lightblue", opacity=0.5)

# Add node labels
for i, coord in enumerate(nodes):
    plotter.add_point_labels(
        [coord],
        [f"{i}"],               # node id
        font_size=14,
        text_color="black",
        point_color="red",
        point_size=12,
        shape=None              # no background box
    )

plotter.add_axes()
plotter.background_color = "white"
plotter.show_grid()

plotter.show(title="Tetrahedron with Node IDs")
```
